```python
from math import sqrt
from numpy import concatenate, split
import numpy as np
from matplotlib import pyplot
from pandas import read_csv
from pandas import DataFrame
from pandas import concat
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_squared_error
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
```

```python
def load_data(filename):
    dataset = read_csv(filename, index_col='index')
    dataset.drop('day', axis=1, inplace=True)
    dataset.columns = ['prec', 'min_t', 'max_t', 'runoff']
    dataset.index.name = 'index'
    return dataset

# convert series to supervised learning
def series_to_supervised(data, timestep):
    cols, names = list(), list()
    for i in range(timestep, -1, -1):
        cols.append(data.shift(i))
        if i == 0:
            names += [('%s(t)' % (col_names)) for col_names in dataset.columns]
        else:
            names += [('%s(t-%d)' % (col_names, i)) for col_names in dataset.columns]
    agg = concat(cols, axis=1)
    agg.columns = names
    agg = agg[timestep:-timestep]
    return agg
```

```python
timestep = 2

dataset = load_data("./KANDERmeteo.csv")
runoff = dataset['runoff'].values
max_runoff, min_runoff = np.max(runoff), np.min(runoff)
values = dataset.values
values.astype('float32')
scaler = MinMaxScaler(feature_range=(0,1))
scaled = scaler.fit_transform(values)
reframed = series_to_supervised(DataFrame(scaled),timestep)
print(reframed.head())

values = reframed.values
test_percentage = 0.2
validation_percentage = 0.1
print([round(values.shape[0]*(1-test_percentage-validation_percentage)),round(values.shape[0]*(1-test_percentage))])
train, validation, test = split(values,[round(values.shape[0]*(1-test_percentage-validation_percentage)),round(values.shape[0]*(1-test_percentage))])
# split into input and outputs
train_X, train_y = train[:, :-1], train[:, -1]
validation_X, validation_y = validation[:, :-1], validation[:, -1]
test_X, test_y = test[:, :-1], test[:, -1]
# reshape input to be 3D [samples, timesteps, features]
train_X = train_X.reshape((train_X.shape[0], 1, train_X.shape[1]))
validation_X = validation_X.reshape((validation_X.shape[0], 1, validation_X.shape[1]))
test_X = test_X.reshape((test_X.shape[0], 1, test_X.shape[1]))
print(train_X.shape, train_y.shape, test_X.shape, test_y.shape)
```

```python
# design network
model = Sequential()
model.add(Dense(train_X.shape[2], input_shape=(train_X.shape[1], train_X.shape[2])))
model.add(LSTM(75))
model.add(Dense(1))
model.compile(loss='mae', optimizer='adam')
# fit network
history = model.fit(train_X, train_y, epochs=10000, batch_size=5000, validation_data=(validation_X, validation_y), verbose=2, shuffle=False)
# plot history
pyplot.plot(history.history['loss'], label='train')
pyplot.plot(history.history['val_loss'], label='validation')
pyplot.legend()
pyplot.show()
```

```python
# make a prediction
yhat = model.predict(test_X)
inv_yhat = yhat * (max_runoff - min_runoff) + min_runoff
# invert scaling for actual
inv_y = test_y * (max_runoff - min_runoff) + min_runoff
# calculate RMSE
pyplot.plot(inv_y, label='obs')
pyplot.plot(inv_yhat, label='pred')
pyplot.legend()
pyplot.show()
rmse = sqrt(mean_squared_error(inv_y, inv_yhat))
print('Test RMSE: %.3f' % rmse)
```
